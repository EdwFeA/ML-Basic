## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
num = input("Введите целое 5 значное число:") # Пользователь вводит число
if len(num) == 5 and num.isdigit(): # Проверка на верное количество символов через len, а также, чтобы они являлись числом через isdigit
    num_1 = num[0] # Первая цифра
    num_2 = num[1:4] # Центральные цифры
    num_3 = num[4] # Последняя цифра
    central = num_2[::-1] # Разворот центра
    
    print(num_1 + central + num_3) # Вывод разворота
else:
    print("Вы ввели не целое 5 значное число, пожалуйста, повторите")
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
days = int(input("Введите сколько дней осталось до отпуска: "))
days -= 1 # Последний день - это отпуск, который не считается

weekends = (days // 7) * 2 # Каждые 7 дней - это всегда 2 выходных
remains = days % 7 # Остаток, который не дотянул до полной недели

if remains >= 5: # Если в остатке есть суббота
    weekends += 1

if remains == 6: # Если в остатке есть суббота и воскресенье
    weekends += 1

print(weekends)
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
choco_l = int(input("Введите длину шоколада: "))
choco_w = int(input("Введите ширину шоколада: "))
choco_s = int(input("Введите размер куска шоколада в дольках, который хотите отломить: "))

total = choco_l * choco_w # Подсчет общего количества долек в плитке 

if choco_s >= total: # Нельзя отломить больше, чем вся шоколадка
    print(False)

if (choco_s % choco_l == 0) or (choco_s % choco_w == 0):
    print(True)

else:
    print(False)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
# Подсчет Римских чисел до 300 

num = int(input("Введите число: "))
rome = ""

# Обработка сотен (C), максимум 3 раза
if num >= 100:
    rome += "C"  # Добавляем C (100)
    num -= 100   # Вычитаем 100
if num >= 100:
    rome += "C"
    num -= 100
if num >= 100:
    rome += "C"
    num -= 100

# Обработка 90 (XC)
if num >= 90:
    rome += "XC"  # Добавляем XC (90)
    num -= 90

# Обработка 50 (L)
if num >= 50:
    rome += "L"   # Добавляем L (50)
    num -= 50

# Обработка 40 (XL)
if num >= 40:
    rome += "XL"  # Добавляем XL (40)
    num -= 40

# Обработка десятков (X), максимум 3 раза
if num >= 10:
    rome += "X"   # Добавляем X (10)
    num -= 10
if num >= 10:
    rome += "X"
    num -= 10
if num >= 10:
    rome += "X"
    num -= 10

# Обработка 9 (IX)
if num == 9:
    rome += "IX"  # Добавляем IX (9)
    num -= 9

# Обработка 5 (V)
if num >= 5:
    rome += "V"   # Добавляем V (5)
    num -= 5

# Обработка 4 (IV)
if num == 4:
    rome += "IV"  # Добавляем IV (4)
    num -= 4

# Обработка единиц (I), максимум 3 раза
if num >= 1:
    rome += "I"   # Добавляем I (1)
    num -= 1
if num >= 1:
    rome += "I"
    num -= 1
if num >= 1:
    rome += "I"
    num -= 1

print("Римское число:", rome)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
num = input("Введите число: ")

if len(num) == 0:
    print(False)
else:
    # Проверяем, что не начинается с минуса
    if num[0] == '-':
        print(False)
    else:
        # Считаем количество точек
        dot_count = num.count('.')
        
        if dot_count > 1:
            print(False)
        else:
            # Убираем одну точку, чтобы проверить остальные символы
            num_without_dot = num.replace('.', '', 1)
            
            # Проверяем, что после удаления точки остались только цифры и num не равна '.'
            if num_without_dot.isdigit() and num != '.':
                print(True)
            else:
                print(False)
```
